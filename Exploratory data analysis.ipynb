# you see this is raw data because it contain number of special charecter and missing value 
#so, its need to be clean before visulaization 
# So the first task is clean the special charecters nad value is in same format
#for removing this special charecter , for catagorical value one method and numerical value other method



```python
Em['Name'] = Em['Name'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
Em
```

```python
Em['Domain'] = Em['Domain'].str.replace(r'\W', '', regex=True)
Em
```

```python
Em['Age'] = Em['Age'].str.replace(r'\W', '', regex=True)
Em
```

```python
Em['Location'] = Em['Location'].str.replace(r'\W', '', regex=True)
Em
```

```python
Em['Salary'] = Em['Salary'].str.replace(r'\W', '', regex=True)
Em
```

```python
Em['Exp'] = Em['Exp'].str.replace(r'\W', '', regex=True)
Em
```

```python
##Now you cleared the special charecter but if you see the numerical coloumn
# you see the types difference , so make them unique by str.extract function

Em
```

```python
Em['Age'] = Em['Age'].str.extract('(\d+)')
Em
```

```python
Em['Exp'] = Em['Exp'].str.extract('(\d+)')
Em
```

# # if you see the table , no special charecter but lot missing value
#work on it

```python
# First do missing value treatment with numerical column
# for this treatment we need use of numpy because you need to perform calculation in array or matrix
#And numpy care about multidimensional array
Em
```

```python
Em["Age"]=Em["Age"].fillna(np.mean(pd.to_numeric(Em["Age"])))
Em
```

```python
#apply same function to salary and Exp
Em["Salary"]=Em["Salary"].fillna(np.mean(pd.to_numeric(Em["Salary"])))
Em
```

```python
Em["Exp"]=Em["Exp"].fillna(np.mean(pd.to_numeric(Em["Exp"])))
Em
```

```python
## now working on missing value of catagorical column
#  check how many null value
Em.isnull().sum()
```

```python
# so Location has 3 null value lets treat with Mode strategy
Em["Location"]=Em["Location"].fillna(Em["Location"].mode()[0])
Em
```

```python
# check the null again
Em.isnull().sum()
```

```python
# check the data type
Em.info()
```

```python
# Did you see Dtype all object by deafault 
# you need convert default to int or catagories by as type()
Em["Name"]=Em["Name"].astype("category")
```

```python
Em.info()
```

```python
Em["Domain"]=Em["Domain"].astype("category")
```

```python
Em["Location"]=Em["Location"].astype("category")
```

```python
#convert to integer
Em["Age"]=Em["Age"].astype("int")
```

```python
Em["Salary"]=Em["Salary"].astype("int")
```

```python
Em["Exp"]=Em["Exp"].astype("int")
```
